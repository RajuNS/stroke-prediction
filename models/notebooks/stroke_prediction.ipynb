## Outlier detection and handling

### Subtask:
Identify and understand the potential impact of outliers in numerical features. Decide on appropriate strategies for handling them if necessary (e.g., capping, transformation, removal).


**Reasoning**:
Identify the relevant numerical columns, calculate the IQR, and determine the number of outliers for each using the 1.5 * IQR rule.



```python
numerical_cols = ['age', 'avg_glucose_level', 'bmi']

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers_lower = df[df[col] < lower_bound]
    outliers_upper = df[df[col] > upper_bound]

    num_outliers_lower = outliers_lower.shape[0]
    num_outliers_upper = outliers_upper.shape[0]
    total_outliers = num_outliers_lower + num_outliers_upper

    print(f"Column: {col}")
    print(f"  IQR: {IQR:.2f}")
    print(f"  Lower Bound: {lower_bound:.2f}")
    print(f"  Upper Bound: {upper_bound:.2f}")
    print(f"  Number of outliers below lower bound: {num_outliers_lower}")
    print(f"  Number of outliers above upper bound: {num_outliers_upper}")
    print(f"  Total number of outliers: {total_outliers}")
    print("-" * 30)
```

## Data Preprocessing

### Subtask: Handle outliers

**Reasoning**:
Based on the EDA, 'avg_glucose_level' and 'bmi' have outliers. I will use capping to handle these outliers by setting values below the lower bound and above the upper bound to the respective bound values.

```python
# Handle outliers in 'avg_glucose_level' and 'bmi' using capping (IQR method)
numerical_cols_outliers = ['avg_glucose_level', 'bmi']

for col in numerical_cols_outliers:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Apply capping
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

print("Outliers handled for 'avg_glucose_level' and 'bmi' using capping.")

# Verify by checking for outliers again
for col in numerical_cols_outliers:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers_lower = df[df[col] < lower_bound]
    outliers_upper = df[df[col] > upper_bound]

    num_outliers_lower = outliers_lower.shape[0]
    num_outliers_upper = outliers_upper.shape[0]
    total_outliers = num_outliers_lower + num_outliers_upper

    print(f"Column: {col}")
    print(f"  Number of outliers below lower bound after capping: {num_outliers_lower}")
    print(f"  Number of outliers above upper bound after capping: {num_outliers_upper}")
    print(f"  Total number of outliers after capping: {total_outliers}")
    print("-" * 30)
```

### Subtask: Address class imbalance

**Reasoning**:
The target variable 'stroke' is highly imbalanced. To address this, I will use the Synthetic Minority Over-sampling Technique (SMOTE) to oversample the minority class ('stroke' = 1).

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Separate features (X) and target (y)
X = df.drop('stroke', axis=1)
y = df['stroke']

# Identify categorical and numerical columns (excluding 'id')
categorical_cols = X.select_dtypes(include='object').columns.tolist()
numerical_cols = X.select_dtypes(include=np.number).columns.tolist()
if 'id' in numerical_cols:
    numerical_cols.remove('id')

# Apply One-Hot Encoding to categorical features
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
X_categorical_encoded = encoder.fit_transform(X[categorical_cols])
X_categorical_encoded_df = pd.DataFrame(X_categorical_encoded, columns=encoder.get_feature_names_out(categorical_cols))

# Combine encoded categorical features with numerical features
X_processed = pd.concat([X[numerical_cols], X_categorical_encoded_df], axis=1)


# Apply SMOTE to the processed data
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_processed, y)

print("Original dataset shape:", X_processed.shape, y.shape)
print("Resampled dataset shape:", X_resampled.shape, y_resampled.shape)
print("Value counts of target variable after SMOTE:")
display(y_resampled.value_counts())
```

## Model selection and training

### Subtask:
Choose appropriate models for binary classification, train them on the preprocessed and engineered data, and evaluate their performance using suitable metrics for imbalanced datasets.


**Reasoning**:
Split the data, train classification models, and evaluate their performance using appropriate metrics.



```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

# 1. Split the resampled data
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled)

# 2. Import and Initialize the chosen models
models = {
    "Logistic Regression": LogisticRegression(random_state=42, solver='liblinear'),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "LightGBM": LGBMClassifier(random_state=42)
}

results = {}

for name, model in models.items():
    print(f"Training {name}...")
    # 4. Train each model
    model.fit(X_train, y_train)

    # 5. Make predictions
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1] # Probability of the positive class

    # 6. Evaluate performance using appropriate metrics
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)

    results[name] = {
        "Precision": precision,
        "Recall": recall,
        "F1-score": f1,
        "AUC": auc
    }

    # Print metrics
    print(f"--- {name} Performance ---")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-score: {f1:.4f}")
    print(f"AUC: {auc:.4f}")
    print("-" * 30)

# Optional: Display results summary
print("\n--- Model Performance Summary ---")
display(pd.DataFrame(results).T)
```
